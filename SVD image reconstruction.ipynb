```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

impath = 'D:/bild_input.jpg'

k = 7 # number of kept principal components
noise = 40 # amplitude of the noise

A = Image.open(impath).convert('L') # load image and convert to grayscale
A = np.double(np.asarray(A))

fig = plt.figure(figsize=(10, 10))
plt.imshow(A, cmap='gray')
plt.title('Original image')
plt.axis('off')
plt.show()
```

```python
noisypic = A + np.random.normal(0, noise, A.shape)
fig = plt.figure(figsize=(10, 10))
plt.imshow(noisypic, cmap='gray')
plt.title('Noisy image')
plt.axis('off')
plt.show()
```

```python
[U,s,V] = np.linalg.svd(noisypic, full_matrices=False)
truncs = s
s[k:] = 0
reconsA = U.dot(np.diag(truncs)).dot(V)

fig = plt.figure(figsize=(10, 10))
plt.imshow(reconsA, cmap='gray')
plt.title('Reconstructed image')
plt.axis('off')
plt.show()
```
